```python
# =========================
# INSTALL (Colab only)
# =========================
!pip install flask pyngrok tensorflow pillow numpy

# =========================
# IMPORTS
# =========================
from flask import Flask, request, jsonify
from pyngrok import ngrok
import tensorflow as tf
import numpy as np
from PIL import Image

# =========================
# LOAD MODEL
# =========================
model = tf.keras.models.load_model("model.keras")

# =========================
# CLASSES
# =========================
classes = [
    "Avian Influenza",
    "Coccidiosis",
    "Coryza",
    "Fowl Cholera",
    "Fowl Pox",
    "Healthy",
    "Infectious Bronchitis",
    "Mareks",
    "Mycoplasma",
    "NewCastle"
]

# =========================
# DISEASE INFO (FULL)
# =========================
info = {
    "Avian Influenza": {
        "symptoms": "Swelling, breathing difficulty, sudden death",
        "remedy": "Isolate birds, disinfect, no cure"
    },
    "Coccidiosis": {
        "symptoms": "Bloody stool, weakness, weight loss",
        "remedy": "Use anticoccidial drugs"
    },
    "Coryza": {
        "symptoms": "Swollen face, nasal discharge, sneezing",
        "remedy": "Antibiotics and isolation"
    },
    "Fowl Cholera": {
        "symptoms": "Diarrhea, weakness, sudden death",
        "remedy": "Antibiotics and vaccination"
    },
    "Fowl Pox": {
        "symptoms": "Skin lesions, scabs on comb",
        "remedy": "Vaccination"
    },
    "Healthy": {
        "symptoms": "No visible symptoms",
        "remedy": "No treatment needed"
    },
    "Infectious Bronchitis": {
        "symptoms": "Coughing, sneezing, nasal discharge",
        "remedy": "Vaccination"
    },
    "Mareks": {
        "symptoms": "Paralysis, weight loss, tumors",
        "remedy": "No cure, vaccination prevention"
    },
    "Mycoplasma": {
        "symptoms": "Respiratory issues, coughing, nasal discharge",
        "remedy": "Antibiotics"
    },
    "NewCastle": {
        "symptoms": "Twisted neck, paralysis, green diarrhea",
        "remedy": "Vaccination"
    }
}

# =========================
# APP
# =========================
app = Flask(__name__)

# =========================
# ROUTE
# =========================
@app.route("/predict", methods=["POST"])
def predict():
    try:
        file = request.files['file']

        img = Image.open(file).convert("RGB").resize((224,224))
        img = np.array(img) / 255.0
        img = np.expand_dims(img, axis=0)

        prediction = model.predict(img)
        class_id = int(np.argmax(prediction))
        confidence = float(np.max(prediction))

        disease = classes[class_id]

        return jsonify({
            "status": "success",
            "disease": disease,
            "confidence": round(confidence * 100, 2),
            "symptoms": info[disease]["symptoms"],
            "remedy": info[disease]["remedy"]
        })

    except Exception as e:
        return jsonify({
            "status": "error",
            "message": str(e)
        })

# =========================
# NGROK
# =========================
public_url = ngrok.connect(5000)
print("🔥 PUBLIC URL:", public_url)

# =========================
# RUN
# =========================
app.run(port=5000)
```
